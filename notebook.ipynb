#### Univariate

Buat fungsi untuk mempermudah dalam proses pencarian insight

```python
import matplotlib.pyplot as plt

def plot_category_counts(data, column_name, top_n=None, bottom_n=None, title=None, xlabel=None, ylabel=None,x_ticks=None, colors=['skyblue', 'salmon', 'lightgreen', 'orange', 'purple']):
    """
    Fungsi untuk membuat bar plot berdasarkan data kategorikal dari suatu kolom dalam DataFrame.
    
    Parameters:
    data : pandas DataFrame
        DataFrame yang berisi data.
    column_name : str
        Nama kolom yang akan diproses.
    top_n : int, optional
        Menampilkan n teratas setelah pengurutan (default None).
    bottom_n : int, optional
        Menampilkan n terbawah setelah pengurutan (default None).
    title : str, optional
        Judul grafik (default adalah None, jika None maka akan mengambil nama kolom).
    xlabel : str, optional
        Label sumbu X (default adalah nama kolom).
    ylabel : str, optional
        Label sumbu Y (default adalah 'Jumlah').
    colors : list, optional
        Daftar warna untuk batang bar plot (default adalah 5 warna).
    """
    
    # Menghitung jumlah kategori untuk kolom yang diberikan
    category_counts = data[column_name].value_counts()
    
    # Urutkan data berdasarkan jumlah kategori
    category_counts_sorted = category_counts.sort_values(ascending=False)
    
    # Ambil teratas atau terbawah jika ada parameter
    if top_n:
        category_counts_sorted = category_counts_sorted.head(top_n)
    elif bottom_n:
        category_counts_sorted = category_counts_sorted.tail(bottom_n)

    # Tentukan jumlah warna sesuai dengan jumlah kategori yang ditampilkan
    color_count = len(category_counts_sorted)
    plot_colors = colors[:color_count]  # Pilih warna yang sesuai dengan jumlah kategori

    # Tentukan judul, xlabel, ylabel jika belum ada
    if title is None:
        title = f'Jumlah Berdasrakan {column_name}'
    if xlabel is None:
        xlabel = column_name
    if ylabel is None:
        ylabel = 'Jumlah'
    
    # Plot
    plt.figure(figsize=(10, 4))
    category_counts_sorted.plot(kind='bar', color=plot_colors)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=x_ticks)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

Disini kita akan membua fungsi untuk menganalisis 2 Variabel, Dimana fokus utama kita adalah analisis Mahsiswa yang droput, dengan kata lain kita akan analisis berdasarkan Dropout

```python
def plot_bivariate_stacked_bar(data, groupby_column, dropout_status_column='Dropout', top_n=None, bottom_n=None, title=None, xlabel=None, ylabel=None, x_ticks=0, colors=['skyblue', 'salmon', 'lightgreen', 'orange', 'purple']):
    """
    Fungsi untuk membuat stacked bar chart berdasarkan hubungan antara dua kolom (bivariate analysis).
    
    Parameters:
    data : pandas DataFrame
        DataFrame yang berisi data.
    groupby_column : str
        Nama kolom yang akan digunakan untuk pengelompokan (misalnya 'Gender', 'Department').
    dropout_status_column : str, optional
        Nama kolom status dropout (default 'Dropout').
    top_n : int, optional
        Menampilkan n teratas setelah pengurutan (default None).
    bottom_n : int, optional
        Menampilkan n terbawah setelah pengurutan (default None).
    title : str, optional
        Judul grafik (default adalah None).
    xlabel : str, optional
        Label sumbu X (default adalah nama kolom_1).
    ylabel : str, optional
        Label sumbu Y (default adalah 'Jumlah').
    x_ticks : int, optional
        Sudut rotasi label pada sumbu X (default 0).
    colors : list, optional
        Daftar warna untuk batang bar plot (default adalah 5 warna).
    """
    
    # Agregasi data dengan groupby berdasarkan kolom yang dipilih dan status Dropout
    grouped_data = data.groupby([groupby_column, dropout_status_column]).size().unstack(fill_value=0)
    
    # Mengurutkan data jika diminta
    if top_n:
        grouped_data = grouped_data.head(top_n)
    elif bottom_n:
        grouped_data = grouped_data.tail(bottom_n)
    
    # Tentukan jumlah warna sesuai dengan jumlah kategori
    color_count = len(grouped_data.columns)
    plot_colors = colors[:color_count]  # Pilih warna yang sesuai dengan jumlah kategori

    # Tentukan judul, xlabel, ylabel jika belum ada
    if title is None:
        title = f'Jumlah Berdasarkan {groupby_column} dan {dropout_status_column}'
    if xlabel is None:
        xlabel = groupby_column
    if ylabel is None:
        ylabel = 'Jumlah'
    
    # Plot stacked bar chart
    plt.figure(figsize=(10, 6))
    grouped_data.plot(kind='bar', stacked=True, color=plot_colors)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=x_ticks)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

Untuk memastikan keterhubungan variabel kategorik terhadap Terjadinya Status Dropout, kita akan melihat menggunakan pendekatan lain yang dikenal sebagai Chi-Square.

```python
df_no_enrolled.info()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Daftar kolom kategorikal yang ingin diuji hubungannya dengan Status
obj_columns = df_no_enrolled.select_dtypes(include=['object'])

# Menghapus kolom 'Status' agar tidak termasuk dalam uji
obj_columns_chi = obj_columns.drop('Status', axis=1)

# Melakukan uji Chi-Square untuk setiap kolom kategorikal terhadap Status
for kolom in obj_columns_chi:
    # Membuat tabel kontingensi antara kolom kategorikal dan Status
    tabel_kontingensi = pd.crosstab(df_no_enrolled[kolom], df_no_enrolled['Status'])
    
    # Melakukan uji Chi-Square
    chi2, p, dof, expected = chi2_contingency(tabel_kontingensi)

    # Menampilkan hasil uji Chi-Square
    print(f"Uji Chi-Square untuk kolom '{kolom}':")
    print(f"Nilai Chi-Square: {chi2:.4f}")
    print(f"P-value: {p:.4f}")

    # Menginterpretasikan hasil uji
    if p < 0.05:
        print(f"=> TOLAK hipotesis nol. Terdapat hubungan yang signifikan antara '{kolom}' dan 'Status'.")
    else:
        print(f"=> GAGAL menolak hipotesis nol. Tidak terdapat hubungan yang signifikan antara '{kolom}' dan 'Status'.")

    print("-" * 60)
```

Tampak bahwa, setelah dilakukan analisis keterhubungan antara fitur **Status** terhadap fitur kategorikal lainnya, dapat dilihaht dari hasil P-Value yang dihasilkan dimana jika P < 0.05 maka dapat dikatakan fitur tersebut berkorelasi. **Bersihkan Outlier**

```python
def winsorize_outliers_iqr(df, columns):
    for col in columns:
        # Hitung Q1 dan Q3 untuk menentukan IQR
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        # Tentukan batas bawah dan atas
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Ganti nilai di luar batas dengan nilai batas
        df[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
    
    return df


col_outlier = [
    'Application_order', 'Previous_qualification_grade', 'Admission_grade', 
    'Age_at_enrollment', 'Curricular_units_1st_sem_credited', 'Curricular_units_1st_sem_enrolled', 
    'Curricular_units_1st_sem_evaluations', 'Curricular_units_1st_sem_approved', 
    'Curricular_units_1st_sem_grade', 'Curricular_units_1st_sem_without_evaluations', 
    'Curricular_units_2nd_sem_credited', 'Curricular_units_2nd_sem_enrolled', 
    'Curricular_units_2nd_sem_evaluations', 'Curricular_units_2nd_sem_approved', 
    'Curricular_units_2nd_sem_grade', 'Curricular_units_2nd_sem_without_evaluations'
]

df_cleaned = winsorize_outliers_iqr(df_no_enrolled, col_outlier)
```
